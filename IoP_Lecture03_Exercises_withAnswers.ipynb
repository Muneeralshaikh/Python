```python
# Exercise 1: The Redundant Department of Redundancy.

# Build a function that takes two integer numbers as its input, n1 and n2,
# and prints 'The X Department of Y',
# where X stands for the word 'Redundant' n1 times,
# and Y stands for the word 'Redundancy' n2 times.
# Bear in mind that the arbitrary numbers n1 and n2 might also be zero.

# Here are some examples of inputs and outputs this function might produce:
#
# n1, n2 -> output_string
# 0, 0   -> The Department
# 0, 1   -> The Department of Redundancy 
# 1, 1   -> The Redundant Department of Redundancy
# 2, 0   -> The Redundant Redundant Department 
# 3, 2   -> The Redundant Redundant Redundant Department of Redundancy Redundancy 

# Write a function that prints the output string given n1 and n2.

#################################
# Your code below here

def department(n, m):
    redundant = ''.join([ 'Redundant ' for i in range(n) ])

    ofredundancy = ''
    if m>0:
        ofredundancy += ' of '
        ofredundancy += ' '.join([ 'Redundancy' for i in range(m) ])
            
    out_str = 'The ' + redundant + 'Department' + ofredundancy
    return out_str
```

```python
# Exercise 4: Code poetry
#
# Take the following phrase 

original_phrase = "My name is Ozymandias, King of Kings; Look at my Works, ye Mighty, and despair! Nothing beside remains."

# (it comes from a poem by Percy Shelley.)
#
# Step 1. Write that phrase to a file called descending_phrase.txt. 
#
# Step 2. Run the remove_a_word function below on that phrase
#         so that it removes a random word from the phrase.
#
# Step 3. Write the new phrase to the file, on a new line. 
#
# Step 4. Repeat steps 2 and 3 on the new phrase, removing words
#         and printing the new phrase to the output file,
#         until the sentence has no words left.
#
# Step 5. Close the file. 
#
# Step 6. Open the file in read mode, and print its output.
#         This way you can check if it worked.

#
# Note.   You can create the output and dump to a file or dump line by line. 
#         Writing in one go is harder on your RAM, writing line by line is harder on the HD.
#         In some contexts, either one can be fair game:
#         printing 100 lines per second would be hard for the hard drive,
#         just like printing 100MB of text at once would be hard for the RAM.
#         But for this exercise, both should be equally fast.

import numpy as np

def remove_a_word(phrase):
    phrase_as_list = phrase.split(' ')
    index = np.random.randint(len(phrase_as_list))
    phrase_as_list.pop(index)
    newphrase = ' '.join(phrase_as_list)
    return newphrase

#################################
# Your code below here

# example answer
# Very didactic
phrase = "My name is Ozymandias, King of Kings; Look at my Works, ye Mighty, and despair! Nothing beside remains."

fileout = open("descending_phrase.txt",'w') # open file

fileout.write(phrase + "\n") # write initial phrase

while len(phrase) > 0: # while the phrase still has characters left
    phrase = remove_a_word(phrase) # remove a word and set as 'phrase'
    fileout.write(phrase + "\n") # write phrase to file

fileout.close() # close file


# A more pythonic way 
phrase = "My name is Ozymandias, King of Kings; Look at my Works, ye Mighty, and despair! Nothing beside remains."

with open("descending_phrase.txt",'w') as fileout:
    while phrase:
        fileout.write(phrase + "\n")
        phrase = remove_a_word(phrase)

# Code check:
print(open("descending_phrase.txt").read())
```

```python
# Exercise 5: Make your own writer

# Write your own function that writes to a file.
#
# This function should take three arguments:
# - The name of the file
# - The text to be written on the file
# - A boolean (i.e. True/False) variable
#   for whether the file should be overwritten, in case the file already exists.
#   This variable should have a default value of True.

# After reading the arguments, the function should cover:
#
# Step 1. Try opening the file in read mode ('r').
#         If it can open the file, it means the file already exists.
#
# Step 2. Check the value of the Boolean (i.e. True/False) variable provided by the user
#
# Step 3. If the value of that variable is True, then overwrite the file.
#         Save the text there and print a message saying "File overwritten".
#
# Step 4: If the value of that variable is false, then do not overwrite the file.
#         In that case, just print a message saying "Mission aborted".
#
# Step 5. Now, if the file doesn't exist, trying to open it in read mode
#         will return an error of the IOError kind. 
#
# Step 6. In case the file doesn't exist, catch that IOError,
#         and open the file in write mode ('w').
#
# Step 7. Once the new file is open, you don't need to worry about overwriting, 
#         so simply write the content there, 
#         and print a message saying "Content written to new file".

# This should test your knowledge of if/else, try/except, functions, and reading/writing files.

#################################
# Your code below here

def file_writer(filename, content, overwrite=True):
    try:
        f = open(filename, 'r') # tries to open file to see if it already exists
        f.close()
        
        if overwrite == True: 
            with open(filename, 'w') as f: # writes info
                f.write(content)
            print('File overwritten')
        else:
            print('Mission aborted')
            
    except IOError: # if there is an error, the file doesn't exist, we can create it
        with open(filename, 'w') as f: # write content to file, no need to ask permission to overwrite
            f.write(content)     
        print('Content written to new file')
```
