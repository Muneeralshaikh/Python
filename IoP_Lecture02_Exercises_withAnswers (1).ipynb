```python
# String and list manipulation 1 . 

# With your own code, devise a way to change the words in the phrase so that
# the old words in replace_dict are remapped to the new words. 
#
# Then when the cell is run, it should print 
# "The quick brown fox jumps over the lazy dog" instead of the variable wrong_phrase.
#
# We use the method split below to split the words by spaces.
#
# To join a list with a space in between each element, you can run this:
# new_string = " ".join(old_list)
#
# You might also want to use the variable wrong_list.
#

wrong_phrase = "The slow red fox watches over the friendly dog"

wrong_list = wrong_phrase.split()

replace_dict = {"slow":"quick",
                "red":"brown",
                "friendly":"lazy",
                "watches":"jumps"}

#################################
# Your code below here

new_list = []

for i in wrong_list: 
    
    if i in replace_dict:
        new_list.append(replace_dict[i])
    else:
        new_list.append(i)

print(" ".join(new_list))
```

```python
# Exercise 3. Food allergies! 

foods = {
         "egg sandwich": ["egg"],
         "side salad": ["nuts"],
         "falafel and pita": [],
         "ice cream": ["dairy","nuts"],
         "fish and chips": ["fish"]
        }

# For every food on the menu, determine if it is vegan or vegetarian,
# Also, if it has nuts, produce a warning
#
# Vegan [vg] food will not have: egg, dairy, or fish
# Vegetarian [v] food will not have fish
#
#   
# The output should look like: 
#
'''
Menu:

egg sandwich [v]
side salad [vg] [nuts]
falafel and pita [vg]
ice cream [v] [nuts]
fish and chips
'''
#
# Hint: if you have a list ['a','b','c'] you can query 
#
# if ( 'a' in my_list )
#
# which will return True because 'a' was in the list above.

#################################
# Your code below here


output = 'Menu\n\n'

for key,value in foods.items(): 
    
    output += "%s" % key
    
    if (not "fish" in value) and (not "dairy" in value) and (not "egg" in value):
        output += " [vg]"
    elif not 'fish' in value:
        output += " [v]"
    
    if "nuts" in value:
        output += " [nuts]"

    output += "\n"

print(output)
```

```python
# Exercise 5. British weather
#
# Part of life in the UK is knowing to expect the worst. Of the weather, at least.
# Imagine the two variables below describe:

# The weather during this week
weather  = [ 'cloudy', 'cloudy', 'sunny', 'cloudy', 'rainy', 'rainy', 'sunny']

# and the items you had in your backpack this week
backpack = [ 'nothing', 'water', 'umbrella', 'nothing', 'umbrella', 'water', 'water' ]

# Your task is to:

# 1. Create a list containing all days in the week

# 2. Use the function zip() to iterate over the two lists above,
#    plus this list you have just created.

# 3. If the day is sunny, but you have not brought any water with you.
#    print the following, replacing Sunday by whatever day of the week it is:
"""
Sunday - Always stay hydrated
"""

# 4. If the day is rainy, but you have not brought an umbrella,
#    print the following, replacing Sunday by whatever day of the week it is:
"""
Sunday - Should've brought an umbrella
"""
    
# 5. And if it is a cloudy day,
#    print the following, replacing Sunday by whatever day of the week it is:
"""
Sunday - Another cloudy day
"""
    
# 6. In all other cases,    
#    print the following, replacing Sunday by whatever day of the week it is:
"""
Sunday - No problems today
"""

# Your final output should be:
"""
Sunday - Another cloudy day
Monday - Another cloudy day
Tuesday - Always stay hydrated
Wednesday - Another cloudy day
Thursday - No problems today
Friday - Should've brought an umbrella
Saturday - No problems today
"""

# PS: I know, umbrellas could be used on sunny days too.
#     But in this exercise your umbrella is transparent.

#################################
# Your code below here

days = [ 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday' ]

for d, w, b in zip(days, weather, backpack):   
    if w == 'cloudy':
        print(d, "- Another cloudy day")
    
    elif (w == 'rainy') and (b != 'umbrella'):
        print(d, "- Should've brought an umbrella")
        
    elif (w == 'sunny') and (b != 'water'):
        print(d, "- Always stay hydrated")
        
    else:
        print(d, "- No problems today")
```

```python
# Exercise 6. British Weather, Part II 

# Silly me. There are other ways to avoid getting wet, or to deal with a sunny day.
# Can you include them all in the code you used above?

# The weather during another week
new_weather  = [ 'rainy', 'rainy', 'sunny', 'cloudy', 'rainy', 'sunny', 'sunny']

# and the items you had in your backpack this week
new_backpack = [ 'poncho','water','sunglasses','nothing','umbrella','umbrella','sunglasses' ]

# Your task is to:

# 1. Create a list containing all days in the week

# 2. Use the function zip() to iterate over the two lists above,
#    plus this list you have just created.

# 3. If the day is sunny, and you brought water or sunglasses,
#    or if the day is rainy, and you brought an umbrella or a poncho,
#    print the following, replacing Sunday by whatever day of the week it is:
"""
Sunday - Rain or shine, I'm ready
"""

# 4. If the day is sunny, forgot your water or sunglasses,
#    print the following, replacing Sunday by whatever day of the week it is:
"""
Sunday - Too sunny for me
"""

# 5. If the day is rainy, forgot your umbrella or poncho
#    print the following, replacing Sunday by whatever day of the week it is:
"""
Sunday - Too rainy for me
"""
        
# 6. And if it is a cloudy day,
#    print the following, replacing Sunday by whatever day of the week it is:
"""
Sunday - I'm always ready for cloudy days
"""
    
# Your final output should be:
"""
Sunday - Rain or shine, I'm ready
Monday - Too rainy for me
Tuesday - Rain or shine, I'm ready
Wednesday - I'm always ready for cloudy days
Thursday - Rain or shine, I'm ready
Friday - Too sunny for me
Saturday - Rain or shine, I'm ready
"""

#################################
# Your code below here


# There are many ways to structure this code

days = [ 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday' ]

for d, w, b in zip(days, new_weather, new_backpack):   
    if w == 'cloudy':
        print(d, "- I'm always ready for cloudy days")
    
    elif (w == 'rainy') and not ((b == 'umbrella') or (b == 'poncho')):
        print(d, "- Too rainy for me")
        
    elif (w == 'sunny') and not ((b == 'water') or (b == 'sunglasses')):
        print(d, "- Too sunny for me")
                    
    else:
        print(d, "- Rain or shine, I'm ready")
```
